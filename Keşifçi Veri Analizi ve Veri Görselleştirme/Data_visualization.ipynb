##  <font color="red">Examination of Missing Values</font>

```python
import seaborn as sns
planets= sns.load_dataset("planets")
df=planets.copy()
df.head()
```

```python
# Is there null values?
df.isnull().values.any()
```

```python
# How many in which variable?
df.isnull().sum()
```

```python
#fill null values with 0
df['orbital_period']=df['orbital_period'].fillna(0, inplace=False)
df.isnull().sum()
```

```python
#fill null values with mean
df['mass']=df['mass'].fillna(df.mass.mean() , inplace=False)
df.isnull().sum()
```

```python
#fill all null values with mean
df=df.fillna(df.mean(numeric_only=True) , inplace=False)
df.isnull().sum()
```

```python
df=planets.copy()
df.head()
```

## <font color="red">Diamond Data(Bar Plot, Cat Plot, Histogram and Distribution) </font>

<font color="blue">**Dataset Story** </font>

<font color="green">
price: dolar cinsinde fiyat ($326 – $18,823)

carat: ağırlık (0.2–5.01)

cut: kalite (Fair, Good, Very Good, Premium, Ideal)

color: renk (from J (worst) to D (best))

clarity: temizliği, berraklığı (I1 (worst), SI2, SI1, VS2, VS1, VVS2, VVS1, IF (best))

x: length in mm (0–10.74)

y: width in mm (0–58.9)

z: depth in mm (0–31.8)

depth: toplam derinlik yüzdesi = z / mean(x, y) = 2 * z / (x + y) (43–79)

table: elmasın en geniş noktasına göre genişliği (43–95) </font>
